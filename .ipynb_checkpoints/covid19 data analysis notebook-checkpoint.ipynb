### Let's Import the modules 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

### importing covid19 dataset
importing "Covid19_Confirmed_dataset.csv" from "./Dataset" folder. 


```python
corona_dataset_csv = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
corona_dataset_csv.head()
```

```python
#shape of the dataframe
corona_dataset_csv.shape
```

### Delete the useless columns

```python
corona_dataset_csv.drop(["Lat","Long"],axis=1,inplace = True)
```

```python
corona_dataset_csv.head(10)
```

###  Aggregating the rows by the country

```python
corona_dataset_aggregad = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregad.head(10)
```

```python
corona_dataset_aggregad.shape
```

### Visualizing data related to a country for example China
visualization always helps for better understanding of our data.

```python
corona_dataset_aggregad.loc["China"].plot()
corona_dataset_aggregad.loc["India"].plot()
corona_dataset_aggregad.loc["Italy"].plot()
plt.legend()
```

###  Calculating a good measure 
we need to find a good measure reperestend as a number, describing the spread of the virus in a country. 

```python
corona_dataset_aggregad.loc['India'].plot()
```

```python
corona_dataset_aggregad.loc["China"][:3].plot()
```

###  caculating the first derivative of the curve

```python
corona_dataset_aggregad.loc["India"].diff().plot()
```

### find maxmimum infection rate for India

```python
corona_dataset_aggregad.loc["India"].diff().max()
```

```python
corona_dataset_aggregad.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregad.loc["China"].diff().max()
```

### find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregad.index)
max_infection_rate = []
for c in countries:
    max_infection_rate.append(corona_dataset_aggregad.loc[c].diff().max())
corona_dataset_aggregad["max_infection_rate"] = max_infection_rate
```

```python
corona_dataset_aggregad.head()
```

### create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_aggregad["max_infection_rate"])
```

```python
corona_data.head()
```
